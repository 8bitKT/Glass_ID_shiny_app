```r
library(shiny)
library(ggplot2)

dataset <- diamonds

pageWithSidebar(

  headerPanel("Diamonds Explorer"),
  
  sidebarPanel(

    sliderInput('sampleSize', 'Sample Size', min=1, max=nrow(dataset),
                value=min(1000, nrow(dataset)), step=500, round=0),
    
    selectInput('x', 'X', names(dataset)),
    selectInput('y', 'Y', names(dataset), names(dataset)[[2]]),
    selectInput('color', 'Color', c('None', names(dataset))),
    
    checkboxInput('jitter', 'Jitter'),
    checkboxInput('smooth', 'Smooth'),
    
    selectInput('facet_row', 'Facet Row', c(None='.', names(dataset))),
    selectInput('facet_col', 'Facet Column', c(None='.', names(dataset)))
  ),

  mainPanel(
    plotOutput('plot')
  )
)
```

```r
iglass_freq <- iris
head(iris)

shinyApp(
    
ui =  fluidPage(
    selectInput(
    
    column(6, 
      checkboxGroupInput("checkGroup", 
        label = h3("Glass Types"),
        choices = list("1: building_windows_float_processed" = 1,
                       "2:building_windows_non_float_processed" = 2, 
                       "3:vehicle_windows_float_processed" = 3,
                       "5:containers" = 4,
                       "6:tableware" = 5,
                       "7:headlamps" = 6), selected = 1)) ),
          
            splitLayout(cellWidths=40,
                plotOutput("iglass_freq", width=40, height=40))),

server = function(input, output){
    output$iglass_freq = renderPlot({
        geom_density(iglass_freq[,input$choices])
    })
}
    
    
)



```

```r
library(shiny)

# Define UI for application that draws a histogram
shinyUI(fluidPage(

  # Application title
  titlePanel("Hello Shiny!"),

  # Sidebar with a slider input for the number of bins
  sidebarLayout(
    sidebarPanel(
      sliderInput("bins",
                  "Number of bins:",
                  min = 1,
                  max = 50,
                  value = 30)
    ),

    # Show a plot of the generated distribution
    mainPanel(
      plotOutput("distPlot")
    )
  )
))
```
